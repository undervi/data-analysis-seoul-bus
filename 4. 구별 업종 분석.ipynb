#### [4.2.2] 원하는 데이터 형태로 저장하기

* head, tail 등을 살펴본 결과를 토대로 header 값을 적절히 지정하여 데이터를 다시 불러오고 필요한 컬럼만 저장하세요

```python
#read_csv 재실행. 단, header값을 적절히 지정
seoul_business = pd.read_csv('./csv/4. business_type.csv', sep=',', header=2)
seoul_business
```

```python
# 다시 불러온 데이터프레임에서 여러분 각자가 사용하고자 하는 컬럼만 가져오세요.
seoul_business = seoul_business.iloc[:, [0, 1, 5, 13, 17, 19, 21]]
```

```python
# 여러분 각자가 사용하고자 하는 형태로 데이터를 정리합니다. 
# 예를 들어, 동 컬럼의 값이 소계인 항목들만 골라내 보겠습니다. 
# 위와 같이 동 컬럼의 소계를 구하면 자치구 별 값이 되므로 이후에 동 컬럼은 삭제하겠습니다. 결과적으로 자치구 별 값을 얻게 됩니다.
seoul_business = seoul_business.loc[seoul_business['동']=='소계']
seoul_business.rename(columns={'종사자수.1': '한식일반음식점업 종사자수', '종사자수.5': '커피전문점 종사자수', '종사자수.7': '기타주점업 종사자수', '종사자수.8': '일반교과학원 종사자수'
                              , '종사자수.9': '한식육류요리전문점 종사자수'}, inplace=True)
seoul_business
```

```python
# seoul_business = seoul_business.drop('동', axis=1, inplace=True)
seoul_business
```

```python
# 위에서 여러분 각자가 사용하고자 하는 형태로 정리한 해당 데이터프레임을 csv 파일('df_seoul_business.csv')로 저장하세요.
seoul_business.to_csv('csv/df_seoul_business.csv', index=False)
```
